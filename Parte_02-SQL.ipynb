#PREGUNTAS

**Pregunta 1**: La gerencia quiere conocer todos los libros disponibles en el catálogo. Muestra el título, autor y precio de todos los libros.

```python
%%sql

SELECT titulo,autor,precio FROM libros
```

**Pregunta 2**: El departamento de marketing quiere hacer una promoción con libros económicos.Muestra todos los libros que tengan un precio menor o igual a 25 euros.

```python
%%sql

SELECT * FROM libros WHERE precio <= 25
```

**Pregunta 3:** Un cliente busca libros de un autor específico.Muestra todos los libros escritos por 'Gabriel García Márquez'.

```python
%%sql

SELECT * FROM libros WHERE autor = "Gabriel García Márquez"
```

**Pregunta 4:**  El equipo de ventas quiere ver los libros ordenados por precio para facilitar recomendaciones. Lista todos los libros ordenados por precio de menor a mayor.

```python
%%sql

SELECT * FROM libros ORDER BY precio ASC
```


**Pregunta 5:**  El gerente de inventario necesita identificar qué libros tienen mayor stock.Muestra los libros ordenados por cantidad en stock de mayor a menor.

```python
%%sql

SELECT * FROM libros ORDER BY stock DESC
```

**Pregunta 6**: La dirección quiere saber cuántos clientes hay registrados en cada ciudad. Muestra cada ciudad y el número de clientes que tiene.

```python
%%sql

SELECT ciudad,count(*) as "numero de clientes" FROM clientes group by ciudad
```

**Pregunta 7:** Finanzas necesita conocer el total de ventas por cliente.
Muestra el nombre de cada cliente y el total de dinero que ha gastado en compras.

```python
%%sql

SELECT nombre,SUM(total) as "Total de dinero gastado"FROM clientes C INNER JOIN ventas V ON c.id_cliente = v.id_cliente group by C.id_cliente
```

**Pregunta 8:** El área de análisis necesita ver las ventas junto con la información de los clientes.Muestra las ventas indicando el nombre del cliente, la fecha de venta y el total de cada venta.

```python
%%sql

SELECT C.nombre,v.fecha_venta,v.total FROM clientes C INNER JOIN ventas V ON c.id_cliente = v.id_cliente
```

**Pregunta 9:** El departamento de reporting necesita un informe completo de ventas con detalles de libros.Muestra las ventas incluyendo el nombre del cliente, el título del libro vendido y el total de la venta.

```python
%%sql

SELECT C.nombre,l.titulo,v.total FROM clientes C INNER JOIN ventas V ON c.id_cliente = v.id_cliente INNER JOIN libros l ON l.id_libro = v.id_libro
```

**Pregunta 10:** La gerencia quiere identificar las ventas más grandes realizadas en Madrid. Muestra las ventas de clientes de Madrid con un total mayor a 30 euros, incluyendo el nombre del cliente y el total, ordenadas por total de mayor a menor.

```python
%%sql

SELECT C.nombre,v.total FROM clientes C INNER JOIN ventas V ON c.id_cliente = v.id_cliente WHERE total > 30 ORDER BY total DESC
```
